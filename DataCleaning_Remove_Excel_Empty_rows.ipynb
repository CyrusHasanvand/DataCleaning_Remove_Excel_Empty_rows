```python
import pandas as pd
```

```python
# Load the Excel file (first sheet by default)
df = pd.read_excel("MemWords.xlsx")
```

```python
df.head()
```

```python
df.tail()
```

```python
# Drop all rows where all columns are empty (i.e., full empty rows)
df_cleaned = df.dropna(how='all')
```

```python
df_cleaned.head()
```

```python
df_cleaned.tail()
```

```python
# Optional: Reset row index 
# I think holding the previous index line may be important in some conditional situation, but we can siply reform them as:
df_cleaned.reset_index(drop=True, inplace=True)
```

```python
df_cleaned.head()
```

```python
df_cleaned.tail()
```

```python
# Save the cleaned version back to Excel
df_cleaned.to_excel("MemWords_Cleaned.xlsx", index=False)

print("Done! Cleaned file saved as MemWords_Cleaned.xlsx")
```
